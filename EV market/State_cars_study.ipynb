```python
# Importing Important Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
df =pd.read_csv("states_data_car.csv")
df.head(15)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x=df["subsidy"])
plt.title('Count plot of subsidy')
plt.xlabel('Subsidy')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x=df["road tax"])
plt.title('Count plot of road tax')
plt.xlabel('road tax')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(25, 7))
sns.set_theme(style="whitegrid")
sns.barplot('state', 'petrol', data=df, color = 'salmon')
plt.title('Petrol Price in Different States' , fontsize = 20)
plt.ylabel('Petrol Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(25, 7))
sns.set_theme(style="whitegrid")
sns.barplot('state', 'diesel', data=df, color = 'salmon')
plt.title('Diesel Price in Different States' , fontsize = 20)
plt.ylabel('Diesel Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```
